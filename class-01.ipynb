## Tools

```python
from agents import Agent, Runner, function_tool
from agents import enable_verbose_stdout_logging

enable_verbose_stdout_logging()


@function_tool
def weather(city: str)-> str:
    return f"The weather in the {city} is rainy."

agent = Agent(name="Assistant", instructions="You are a helpful assistant.", tools=[weather])


# result = Runner.run_sync(agent, "Capital of Pakistan")

result = await Runner.run(
            agent,
            input='What is the weather in Karachi',
            run_config=config
        )
print(result.final_output)
print()

print(result)
```

```python
import asyncio
from agents import Agent, Runner, function_tool, enable_verbose_stdout_logging
from config import config  # <--- Make sure this line exists!

enable_verbose_stdout_logging()
config.tracing_disabled = False


@function_tool
def weather(city: str) -> str:
    return f"The weather in {city} is rainy."


# Define the agent with tool
agent = Agent(
    name="Assistant",
    instructions="You are a helpful assistant.",
    tools=[weather]
)


async def stream_agent():
    result = await Runner.run_streamed(
        agent,
        input='What is the weather in Karachi',
        run_config=config
    )

    async for event in result.stream_events():
        print(f"\n[Event]: {event}\n")


# Run the async event loop
import nest_asyncio
import asyncio

nest_asyncio.apply()  # Patch the running event loop if needed

asyncio.run(stream_agent())  # This works in Jupyter/Chainlit/etc.
```
